```python
n = 3
m = 2
array = [1, 1, 5, 3]
like = {3, 1}
dislike = {5, 7}
```

```python
# BLOCK 2
happiness_score = 0
print(array)
for num in array:
    if num in like:
        happiness_score += 1
    elif num in dislike:
        happiness_score -= 1
    print(happiness_score)
print(happiness_score)
```
