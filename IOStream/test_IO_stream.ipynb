```python
# Writing to a file
with open("iostream1.txt", "w", encoding="utf-8") as f:
    f.write("Hello, file stream!")

# Reading from a file
with open("iostream1.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```
